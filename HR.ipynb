```python
# Basic cleaning
df.columns = df.columns.str.strip().str.lower()
```

```python
df.columns
```

```python
# Handle missing values
df['monthly income'].fillna(df['monthly income'].median(), inplace=True)
df['job satisfaction'].fillna(df['job satisfaction'].mode()[0], inplace=True)
```

```python
# Convert categorical fields
df['attrition_flag'] = df['attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Loyalty score (example feature)
df['loyalty_score'] = df['years at company'] * df['job satisfaction']
```

```python
df.head()
```

```python
# Dimension Table: Employee
dim_employee = df[['employee number','age','gender','department']].drop_duplicates()

# Fact Table: HR Metrics
fact_hr = df[['employee number','monthly income','job satisfaction',
              'attrition_flag','loyalty_score','performance rating']]
```

```python
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# SQL Server details
server = r'NIKHILC\SQLEXPRESS'
database = 'HR_Analytics'

# URL encode the driver name (avoid issues with spaces)
driver = quote_plus("ODBC Driver 17 for SQL Server")

# Create SQLAlchemy engine using Windows Authentication
engine = create_engine(
    f"mssql+pyodbc://{server}/{database}?driver={driver}&trusted_connection=yes"
)

# Example DataFrames (replace with your actual data)
# dim_employee = pd.DataFrame(...)
# fact_hr = pd.DataFrame(...)

# Write DataFrames to SQL Server
dim_employee.to_sql('dim_employee', engine, if_exists='replace', index=False)
fact_hr.to_sql('fact_hr', engine, if_exists='replace', index=False)

print("Tables written successfully!")
```
